```python
import pandas as pd

movies = pd.read_csv("movies.csv")

movies.info()
```

```python
grp_genre = movies.groupby("Genre")

genre_count = []

for grp, grp_info in grp_genre:
    cat_dict = {}
    grp_gen = grp_genre.get_group(grp)

    for index,row in grp_gen.iterrows(): # to read the data inside those row and column
        if row["Genre"] not in cat_dict.keys():
            cat_dict[row["Genre"]] = [row["Film"]]
            
        else:
            if row["Film"] not in cat_dict.values():
                cat_dict[row["Genre"]].append(row["Film"])
    genre_count.append(cat_dict)


genre_count
```

```python
for row in genre_count:
    for  key,value in row.items():
        count = 0
        if isinstance(value, list): # To check whether the value is list form.
            for v in value:
                count +=1
            row[key] = count 
genre_count
```

```python
grp_genre = movies.groupby("Genre")

audience_score = []

for grp, grp_info in grp_genre:
    cat_dict = {}
    grp_gen = grp_genre.get_group(grp)

    for index,row in grp_gen.iterrows(): # to read the data inside those row and column
        if row["Genre"] not in cat_dict.keys():
            cat_dict[row["Genre"]] = [row["Audience score %"]]
            
        else:
            if row["Audience score %"] not in cat_dict.values():
                cat_dict[row["Genre"]].append(row["Audience score %"])
    audience_score.append(cat_dict)


audience_score
```

```python
for row in audience_score:
    for  key,value in row.items():
        average = 0 
        if isinstance(value, list): # To check whether the value is list form.
            average = round(sum(value) / len(value),2)
            row[key] = average
            
audience_score
```
